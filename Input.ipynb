```python
import os

# Trích xuất mã chứng khoán từ DataFrame
if not all_companies_df.empty and 'symbol' in all_companies_df.columns:
    stock_symbols = all_companies_df['symbol'].tolist()
    print(f"Successfully retrieved {len(stock_symbols)} stock symbols from all_companies_df.")
elif not all_companies_df.empty and 'ticker' in all_companies_df.columns:
    stock_symbols = all_companies_df['ticker'].tolist()
    print(f"Successfully retrieved {len(stock_symbols)} stock symbols using 'ticker' column from all_companies_df.")
else:
    print("Could not retrieve stock symbols. Listing DataFrame does not contain 'symbol' or 'ticker' column, or is empty.")
    stock_symbols = []

print(f"Stock symbols to process: {stock_symbols[:10]}... (showing first 10)")

start_date = '2021-01-01'
end_date = '2025-12-31'
output_dir = os.path.join(os.getcwd(), 'output')
output_filename = 'historical_stock_data_full.csv'

all_ohlcv_data = []

if stock_symbols:
    for symbol in stock_symbols:
        print(f"Fetching historical data for {symbol}...")
        try:
            df_symbol = client.get_stock_historical(symbol, start_date, end_date)
            if not df_symbol.empty:
                df_symbol['symbol'] = symbol  # Add a column to identify the stock
                all_ohlcv_data.append(df_symbol)
            else:
                print(f"No data found for {symbol}")
        except Exception as e:
            print(f"Error fetching data for {symbol}: {e}")
else:
    print("No stock symbols available to fetch data for.")

# Ghép tất cả các DataFrame lại với nhau thành một DataFrame duy nhất.
if all_ohlcv_data:
    combined_df = pd.concat(all_ohlcv_data, ignore_index=True)
    display("Combined historical data head:", combined_df.head())
    display("Combined historical data tail:", combined_df.tail())

    # Đảm bảo thư mục đầu ra tồn tại.
    try:
        os.makedirs(output_dir, exist_ok=True)
    except PermissionError:
        output_dir = os.getcwd()
        print(f"Permission denied creating the requested output directory. Saving to current working directory instead: {output_dir}")
        os.makedirs(output_dir, exist_ok=True)

    # Xác định đường dẫn đầy đủ cho tệp CSV đầu ra.
    output_filepath = os.path.join(output_dir, output_filename)

    # Lưu DataFrame đã kết hợp vào tệp CSV.
    combined_df.to_csv(output_filepath, index=False)
    print(f"Combined historical data for all stocks saved to {output_filepath}")
else:
    print("No historical data was fetched for any of the symbols.")
```
