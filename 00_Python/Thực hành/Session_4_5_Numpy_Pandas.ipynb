
# Numpy


```python
import numpy as np
```

## Cú pháp:
- import numpy as np
- array_name = np.array([elements])


## Ví dụ:

```python
# Tạo mảng NumPy
prices = np.array([100, 101, 102, 103, 104])
print("Mảng giá cổ phiếu:", prices)
```

### Truy cập phần tử:

```python
first_price = prices[0]
print("Giá cổ phiếu đầu tiên:", first_price)
```

### Cắt mảng (slicing):

```python
sliced_prices = prices[1:3]
print("Giá cổ phiếu từ ngày thứ 2 đến thứ 3:", sliced_prices)
```

### Thay đổi kích thước mảng (reshaping):


```python
reshaped_prices = prices.reshape((5, 1))
print("Mảng giá cổ phiếu sau khi thay đổi kích thước:", reshaped_prices)
```

# Phép toán vector hóa (Vectorized operations)
## Cú pháp:
- result = array1 operation array2


```python
# Tính lợi nhuận hàng ngày
daily_returns = (prices[1:] - prices[:-1]) / prices[:-1]
print("Lợi nhuận hàng ngày:", daily_returns)

#Tính tổng của hai mảng:
total_prices = prices + np.array([1, 1, 1, 1, 1])
print("Tổng giá cổ phiếu:", total_prices)

#Tính hiệu của hai mảng:
price_diff = prices - np.array([1, 1, 1, 1, 1])
print("Hiệu giá cổ phiếu:", price_diff)
```

# Broadcasting
## Cú pháp:
- result = array operation scalar


```python
# Broadcasting
adjusted_prices = prices * 1.02
print("Giá cổ phiếu sau khi điều chỉnh:", adjusted_prices)

# Nhân mảng với một số:
scaled_prices = prices * 2
print("Giá cổ phiếu sau khi nhân đôi:", scaled_prices)

#Chia mảng cho một số:
half_prices = prices / 2
print("Giá cổ phiếu sau khi chia đôi:", half_prices)
```

# Hàm toán học và thống kê
## Cú pháp:
- result = np.function_name(array)

```python
# Tính toán các chỉ số thống kê
mean_price = np.mean(prices)
std_price = np.std(prices)

print("Giá trung bình:", mean_price)
print("Độ lệch chuẩn của giá:", std_price)

#Tính giá trị trung bình:
mean_price = np.mean(prices)
print("Giá trung bình:", mean_price)

#Tính độ lệch chuẩn:
std_price = np.std(prices)
print("Độ lệch chuẩn của giá:", std_price)
```

# DataFrame và Series
## Cú pháp:
- Tạo Series: series_name = pd.Series(data)
- Tạo DataFrame: dataframe_name = pd.DataFrame(data)


```python
import pandas as pd
# Tạo Series từ danh sách
s = pd.Series([1, 3, 5, 7, 9])
print("Series từ danh sách:\n", s)

# Tạo DataFrame từ từ điển
data = {'Name': ['John', 'Anna', 'Peter', 'Linda'],
        'Age': [28, 32, 35, 32]}
df = pd.DataFrame(data)
df.at[2, 'Age'] = 'na'
print("DataFrame từ từ điển:\n", df)
```

```python
# Truy cập phần tử trong Series:
first_element = s[0]
print("Phần tử đầu tiên của Series:", first_element)

#Truy cập cột trong DataFrame:
names = df['Name']
print("Cột 'Name' trong DataFrame:", names)

#Truy cập hàng trong DataFrame:
first_row = df.iloc[0]
print("Hàng đầu tiên trong DataFrame:", first_row)
```

# Thao tác dữ liệu
## Cú pháp:
- Lọc dữ liệu: filtered_df = dataframe_name[dataframe_name['column_name'] condition]
- Nhóm và tổng hợp dữ liệu: grouped_df = dataframe_name.groupby('column_name').aggregate_function()


```python
# Tạo DataFrame với 3 hàng và 4 cột: Date, Symbol, Close, Volume
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-02'],
    'Symbol': ['AAPL', 'GOOGL', 'AAPL'],
    'Close': [150.75, 2750.30, 299.50],
    'Volume': [1000000, 1500000, 1200000]
}

df = pd.DataFrame(data)

# Lọc dữ liệu
filtered_df = df[df['Close'] > 100]
print("Dữ liệu sau khi lọc:\n", filtered_df.head())


# Nhóm và tính trung bình chỉ trên các cột số
grouped_df = df.groupby('Date')[['Close', 'Volume']].mean()
print("Dữ liệu sau khi nhóm và tính trung bình:\n", grouped_df)
```

# Ứng dụng của Pandas trong phân tích tài chính
## Cú pháp:
- Điền giá trị thiếu: dataframe_name.fillna(value, inplace=True)
- Loại bỏ giá trị thiếu: dataframe_name.dropna(inplace=True)


```python
import matplotlib.pyplot as plt
# Tính toán lợi nhuận hàng ngày
df['Daily Return'] = df['Close'].pct_change()
print("Lợi nhuận hàng ngày:", df['Daily Return'].head())

# Tính trung bình và độ lệch chuẩn của lợi nhuận hàng ngày
mean_return = df['Daily Return'].mean()
std_return = df['Daily Return'].std()

print("Lợi nhuận trung bình hàng ngày:", mean_return)
print("Độ lệch chuẩn của lợi nhuận hàng ngày:", std_return)

# Trực quan hóa dữ liệu tài chính
df['Close'].plot(title='Giá cổ phiếu theo thời gian')
plt.xlabel('Ngày')
plt.ylabel('Giá đóng cửa')
plt.show()
```

```python
# Tính phần trăm thay đổi chỉ trên các cột số
returns = df[['Close', 'Volume']].pct_change()
print("Phần trăm thay đổi:\n", returns.tail(2))

# Tính ma trận tương quan
correlation_matrix = returns.corr()
print("Ma trận tương quan:\n", correlation_matrix)

# Chuyển đổi cột ngày tháng và đặt làm chỉ số:
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Chuyển đổi tần suất dữ liệu và tính giá trị trung bình chỉ trên các cột số
monthly_data = df.resample('M').mean(numeric_only=True)
print("Dữ liệu theo tháng:\n", monthly_data.head(2))
```

```python
# Tính trung bình động 20 ngày
df['20 Day MA'] = df['Close'].rolling(window=20).mean()

# Tạo tín hiệu mua/bán dựa trên trung bình động
df['Signal'] = np.where(df['Close'] > df['20 Day MA'], 1, 0)

# Hiển thị DataFrame với các cột mới
print(df[['Close', '20 Day MA', 'Signal']].head(2))
```
