```python
import pandas as pd
from agentic_vnstock.client import AgenticVNStock

# Khởi tạo client
client = AgenticVNStock()
```

```python
# 1.1 Giá lịch sử (Sử dụng API Entrade/DNSE)
df_history = client.get_stock_historical('HPG', '2024-01-01', '2024-01-10')
display("Giá lịch sử:", df_history.head())
```

```python
# 1.2 Thông tin tổng quan doanh nghiệp (Sử dụng API Vietcap)
df_overview = client.get_company_overview('FRT')
display("Thông tin tổng quan:", df_overview.head())
```

```python



# 1.3 Chỉ số tài chính (Ví dụ cấu trúc SSI API)
df_ratio = client.get_financial_ratio('HPG')
display("Chỉ số tài chính:", df_ratio.head())

# 1.4 Báo cáo tài chính (Cân đối kế toán, KQ Kinh doanh, Lưu chuyển tiền tệ)
df_financial_report = client.get_financial_report('HPG', report_type='IncomeStatement', period='Quarterly')
display("Báo cáo tài chính (KQKD):", df_financial_report.head())
```

```python
# VNINDEX
df_vnindex = client.get_market_index('VNINDEX', '2024-01-01', '2024-01-10')
display("Chỉ số VNINDEX:", df_vnindex.head())
```

```python
# Lấy giá phái sinh (Ví dụ mã hợp đồng VN30F1M)
# API của Entrade xử lý các mã phái sinh như một mã chứng khoán (symbol) thông thường.
df_future = client.get_derivative_or_cw('VN30F1M', '2024-01-01', '2024-01-10')
display("Phái sinh (VN30F1M):", df_future.head())
```

```python
# Lấy dữ liệu Chứng quyền (Ví dụ mã chứng quyền CHPG2301)
df_cw = client.get_derivative_or_cw('CHPG2301', '2024-01-01', '2024-01-10')
display("Chứng quyền (CHPG2301):", df_cw.head())
```

```python
# Danh sách các chứng chỉ quỹ (Gọi từ API POST của Fmarket)
df_fund = client.get_fund_data()
display("Danh sách Quỹ mở/ETF:", df_fund.head())
```

```python
# Dữ liệu trái phiếu doanh nghiệp/trái phiếu chính phủ (Cần cập nhật endpoint chính xác từ HNX/TCBS)
df_bond = client.get_bond_data()
display("Trái phiếu:", df_bond.head())
```

```python
# Lấy tin tức doanh nghiệp liên quan đến mã chứng khoán (Sử dụng API Vietcap)
df_news = client.get_company_news('HPG')
display("Tin tức HPG:", df_news.head())
```
